```julia
using HorizonSideRobots
using PyPlot; pygui(true)
h, w = 10, 10
r = Robot(h, w)
show!(r)
```

```julia
show(r)
```

```julia
x, y = 0, 0

function check(r, x, y)
    if (abs(x) + abs(y)) % 2 == 0
        putmarker!(r)
    end
end

function get_to_start(r, x, y)
    for i in (1:abs(x))
        move!(r, Sud)
    end
    for i in (1:abs(y))
        move!(r, Ost)
    end
end

function move_till_border(r, side)
    global x
    global y
    check(r, x, y)
    while(!isborder(r, side))
        move!(r, side)
        if(Int(side) % 2 == 0)
            x += Int(side) - 1
        else
            y += Int(side) - 2
        end
        check(r, x, y)
    end
end

println(string(x) * " " * string(y))
move_till_border(r, Sud)
println(string(x) * " " * string(y))
move_till_border(r, West)
println(string(x) * " " * string(y))


while !isborder(r, Nord)
    move_till_border(r, Ost)
    move_till_border(r, West)
    move!(r, Nord)
    x -= 1
end

move_till_border(r, Ost)
move_till_border(r, West)
get_to_start(r, x, y)
```
