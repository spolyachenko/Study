```julia
function inverse_side(side)
    return HorizonSide((Int(side) + 2) % 4)
end

function walk_back(r, path)
    for side in reverse(path)
        move!(r, inverse_side(side))
    end
end

function mark_branch(r, side1, side2)
    path = Vector()
    while(!isborder(r, side1) && !isborder(r, side2))
        move!(r, side1)
        move!(r, side2)
        putmarker!(r)
        push!(path, side1)
        push!(path, side2)
    end
    walk_back(r, path)
end

function put_cross(r)
    putmarker!(r)
    mark_branch(r, Nord, West)
    mark_branch(r, Nord, Ost)
    mark_branch(r, Sud, West)
    mark_branch(r, Sud, Ost)
end

put_cross(r)
show(r)
```
