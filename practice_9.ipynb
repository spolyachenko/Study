```julia
function move_to_side(r, len, side)
    b = false
    for i in (1:len)
        if ismarker(r)
            b = true
            break
        end
        move!(r, side)
    end
    return b
end

function check_rect(r, len)
    b = move_to_side(r, len, Nord)
    if(b)
        return b
    end
    b = move_to_side(r, len, Ost)
    if(b)
        return b
    end
    b = move_to_side(r, len, Sud)
    if(b)
        return b
    end
    b = move_to_side(r, len, West)
    if(b)
        return b
    end
    return b
end

len = 0
b = check_rect(r, len)
while(!b)
    move!(r, Sud)
    move!(r, West)
    len += 2
     b = check_rect(r, len)
end
```
