```julia
using HorizonSideRobots
using PyPlot; pygui(true)
h, w = 10, 11
r = Robot(h, w)
show!(r)
```

```julia
show(r)
```

```julia
function inverse(side)
    return HorizonSide(mod(Int(side) + 2, 4))
end

function go_to(r, direction, start_x, end_x)
    while (end_x != start_x) && (isborder(r, Nord))
        move!(r, direction)
        start_x += Int(direction) - 2
    end
    if(!isborder(r, Nord))
        return true
    end
    return false
end

step = 0
dir = Ost
b = go_to(r, dir, 0, step)
while(!b)
    step += 1
    b = go_to(r, dir, 0, step)
    dir = inverse(dir)
    if b
        break
    end
    b = go_to(r, dir, step, -step)
    dir = inverse(dir)
    if(b)
        break
    end
    go_to(r, dir, -step, 0)
end
```
