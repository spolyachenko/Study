```julia
using HorizonSideRobots
using PyPlot; pygui(true)
h, w = 10, 10
r = Robot(h, w)
show!(r)

function get_to_start(r, x, y)
    side_x = Nord
    if x < 0
        side_x = Sud
    end
    for i in (1:abs(x))
        move!(r, side_x)
    end
    side_y = West
    if y < 0
        side_y = Ost
    end
    for i in (1:abs(y))
        move!(r, side_y)
    end
end
```

```julia
x, y = 0, 0
function move_to_border(f, r, side, dif = 0)
    res = 0
    while !isborder(r, side)
        f(r)
        move!(r, side)
        res += dif
    end
    return res
end
function move_to(f, r, w, side, dif = 0)
    res = 0
    if w <= 0
        return 0
    end
    for i in (1:w-1)
        f(r)
        move!(r, side)
        res += dif
    end
    f(r)
    return res
end

f(a) = a
x += move_to_border(f, r, Sud, 1)
y += move_to_border(f, r, West, -1)
for i in (0:h-2)
    move_to(putmarker!, r, w - i, Ost)
    move_to(f, r, w - i, West)
    move!(r, Nord)
    x -= 1
end
move_to(putmarker!, r, w - h + 1, Ost)
move_to(f, r, w - h + 1, West)
println(x, y)
get_to_start(r, x, y)
```
